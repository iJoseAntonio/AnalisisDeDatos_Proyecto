```python
engine = create_engine(
    "mssql+pyodbc://@localhost/Liga1?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes"
)

ddl = r"""
-- Borrar si existen (en orden para respetar FKs)
IF OBJECT_ID('dbo.Estadisticas_partidos','U') IS NOT NULL DROP TABLE dbo.Estadisticas_partidos;
IF OBJECT_ID('dbo.Partidos','U')              IS NOT NULL DROP TABLE dbo.Partidos;
IF OBJECT_ID('dbo.Equipos','U')               IS NOT NULL DROP TABLE dbo.Equipos;

-- Crear tablas
CREATE TABLE dbo.Equipos (
  equipo_id INT IDENTITY(1,1) PRIMARY KEY,
  nombre    VARCHAR(50) NOT NULL
);

CREATE TABLE dbo.Partidos (
  partido_id           BIGINT IDENTITY(1,1) PRIMARY KEY,
  fecha                DATE NOT NULL,
  equipo_local_id      INT  NOT NULL,
  equipo_visitante_id  INT  NOT NULL,
  goles_local          SMALLINT NOT NULL,
  goles_visitante      SMALLINT NOT NULL,
  url_partido          VARCHAR(300) NULL,
  temporada            INT  NOT NULL,
  CONSTRAINT FK_Partidos_Local  FOREIGN KEY (equipo_local_id)     REFERENCES dbo.Equipos(equipo_id),
  CONSTRAINT FK_Partidos_Visit  FOREIGN KEY (equipo_visitante_id) REFERENCES dbo.Equipos(equipo_id),
  CONSTRAINT UQ_Partidos UNIQUE (temporada, fecha, equipo_local_id, equipo_visitante_id),
  CONSTRAINT CK_Partidos_Goles CHECK (goles_local >= 0 AND goles_visitante >= 0),
  CONSTRAINT CK_Partidos_Equipos CHECK (equipo_local_id <> equipo_visitante_id)
);

CREATE TABLE dbo.Estadisticas_partidos (
  partido_id            BIGINT   NOT NULL,
  equipo_id             INT      NOT NULL,
  es_local              BIT      NOT NULL,    -- 1 local, 0 visitante
  tiros_totales         SMALLINT NULL,
  tiros_porteria        SMALLINT NULL,
  atajadas_portero      SMALLINT NULL,
  pases                 INT      NULL,
  saques_esquina        SMALLINT NULL,
  faltas                SMALLINT NULL,
  tarjetas_amarillas    SMALLINT NULL,
  tarjetas_rojas        SMALLINT NULL,
  CONSTRAINT PK_Estadisticas PRIMARY KEY (partido_id, equipo_id),
  CONSTRAINT FK_Estadisticas_Partido FOREIGN KEY (partido_id) REFERENCES dbo.Partidos(partido_id),
  CONSTRAINT FK_Estadisticas_Equipo  FOREIGN KEY (equipo_id)   REFERENCES dbo.Equipos(equipo_id),
  CONSTRAINT CK_Estadisticas_EsLocal CHECK (es_local IN (0,1))
);
"""

with engine.begin() as conn:
    for stmt in ddl.split(";"):
        s = stmt.strip()
        if s:
            conn.execute(text(s))
```

```python

# Configuración
EXCEL_PATH = r"Liga1_2024_2025.xlsx"   
SHEET_NAME = 'Liga1'   

ENGINE = create_engine(
    "mssql+pyodbc://@localhost/Liga1?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes",
    fast_executemany=True,
)

# Mapeo: columnas del Excel -> columnas SQL (tabla dbo.Estadisticas_partidos)
STATS_LOCAL_XLS_TO_SQL = {
    "Tiros totales_local":           "tiros_totales",
    "Disparos a puerta_local":       "tiros_porteria",
    "Paradas del portero_local":     "atajadas_portero",
    "Pases_local":                   "pases",
    "Saques de esquina_local":       "saques_esquina",
    "Faltas_local":                  "faltas",
    "Tarjetas amarillas_local":      "tarjetas_amarillas",
    "Tarjetas rojas_local":          "tarjetas_rojas",
}
STATS_VISITA_XLS_TO_SQL = {
    "Tiros totales_visitante":       "tiros_totales",
    "Disparos a puerta_visitante":   "tiros_porteria",
    "Paradas del portero_visitante": "atajadas_portero",
    "Pases_visitante":               "pases",
    "Saques de esquina_visitante":   "saques_esquina",
    "Faltas_visitante":              "faltas",
    "Tarjetas amarillas_visitante":  "tarjetas_amarillas",
    "Tarjetas rojas_visitante":      "tarjetas_rojas",
}

BASE_COLS_REQUIRED = {"fecha", "local", "visitante", "marcador", "url"}

# 1) LEER EXCEL 
def leer_hoja_correcta(path, sheet_name=None):
    if sheet_name is not None:
        return pd.read_excel(path, sheet_name=sheet_name)

    xls = pd.ExcelFile(path)
    for sh in xls.sheet_names:
        tmp = pd.read_excel(path, sheet_name=sh, nrows=5)
        cols_norm = {str(c).strip().lower() for c in tmp.columns}
        if BASE_COLS_REQUIRED.issubset(cols_norm):
            return pd.read_excel(path, sheet_name=sh)
    raise ValueError(f"No encontré ninguna hoja con columnas {BASE_COLS_REQUIRED}. Hojas: {xls.sheet_names}")

df = leer_hoja_correcta(EXCEL_PATH, SHEET_NAME)

# Normaliza/renombra columnas base
df = df.rename(columns={
    "fecha": "fecha",
    "local": "equipo_local",
    "visitante": "equipo_visitante",
    "marcador": "marcador",
    "url": "url_partido",
})

# Limpieza mínima
df = df.dropna(subset=["fecha", "equipo_local", "equipo_visitante", "marcador"]).copy()
df["equipo_local"] = df["equipo_local"].astype(str).str.strip()
df["equipo_visitante"] = df["equipo_visitante"].astype(str).str.strip()

# Parse del marcador "A-B" -> goles_local / goles_visitante
mar = df["marcador"].astype(str).str.strip().str.split("-", n=1, expand=True)
df["goles_local"] = pd.to_numeric(mar[0], errors="coerce")
df["goles_visitante"] = pd.to_numeric(mar[1], errors="coerce")
df = df.dropna(subset=["goles_local", "goles_visitante"]).copy()
df["goles_local"] = df["goles_local"].astype(int)
df["goles_visitante"] = df["goles_visitante"].astype(int)

# Temporada = año de la fecha (ajusta si tu definición es distinta)
df["fecha"] = pd.to_datetime(df["fecha"], errors="coerce")
if df["fecha"].isna().any():
    filas_malas = df[df["fecha"].isna()]
    raise ValueError(f"Hay fechas inválidas en el Excel:\n{filas_malas[['equipo_local','equipo_visitante','marcador']].head()}")
df["temporada"] = df["fecha"].dt.year.astype(int)

# Evitar duplicados por la UNIQUE de Partidos
df = df.drop_duplicates(subset=["temporada", "fecha", "equipo_local", "equipo_visitante"]).copy()

# 2) EQUIPOS 
with ENGINE.begin() as conn:
    equipos_uniq = pd.DataFrame({
        "nombre": pd.unique(pd.concat([df["equipo_local"], df["equipo_visitante"]], ignore_index=True))
    })
    equipos_uniq["nombre"] = equipos_uniq["nombre"].astype(str).str.strip()

    existentes = pd.read_sql(text("SELECT equipo_id, nombre FROM dbo.Equipos"), conn)
    existentes["nombre"] = existentes["nombre"].astype(str).str.strip()

    nuevos = equipos_uniq[~equipos_uniq["nombre"].isin(existentes["nombre"])].copy()
    if not nuevos.empty:
        nuevos.to_sql("Equipos", conn, schema="dbo", if_exists="append", index=False)
        print(f"✓ Equipos nuevos insertados: {len(nuevos)}")
    else:
        print("✓ No hay equipos nuevos.")

    # Refrescar mapping equipo -> id
    equipos_db = pd.read_sql(text("SELECT equipo_id, nombre FROM dbo.Equipos"), conn)
    equipos_db["nombre"] = equipos_db["nombre"].astype(str).str.strip()
    mapa_equipo = dict(zip(equipos_db["nombre"], equipos_db["equipo_id"]))

# Mapear FKs
df["equipo_local_id"] = df["equipo_local"].map(mapa_equipo)
df["equipo_visitante_id"] = df["equipo_visitante"].map(mapa_equipo)

faltantes = df[df["equipo_local_id"].isna() | df["equipo_visitante_id"].isna()][["equipo_local","equipo_visitante"]].drop_duplicates()
if not faltantes.empty:
    raise ValueError(f"Hay equipos sin id (revisa tildes/espacios):\n{faltantes}")

# 3) PARTIDOS
partidos_cols = ["fecha","equipo_local_id","equipo_visitante_id","goles_local","goles_visitante","url_partido","temporada"]
partidos_df = df[partidos_cols].copy()

# Para comparar correctamente: usa fechas como 'date' (no datetime) en ambos lados
partidos_df["fecha"] = pd.to_datetime(partidos_df["fecha"]).dt.date

with ENGINE.begin() as conn:
    ya = pd.read_sql(text("""
        SELECT temporada, CAST(fecha AS date) AS fecha, equipo_local_id, equipo_visitante_id
        FROM dbo.Partidos
    """), conn)

# Anti-join para insertar solo nuevos
if not ya.empty:
    key = ["temporada","fecha","equipo_local_id","equipo_visitante_id"]
    candidatos = partidos_df.copy()
    merge = candidatos.merge(ya.drop_duplicates(), on=key, how="left", indicator=True)
    nuevos_partidos = merge[merge["_merge"] == "left_only"][partidos_cols].copy()
else:
    nuevos_partidos = partidos_df.copy()

with ENGINE.begin() as conn:
    if not nuevos_partidos.empty:
        nuevos_partidos.to_sql("Partidos", conn, schema="dbo", if_exists="append", index=False)
        print(f"✓ Partidos insertados: {len(nuevos_partidos)}")
    else:
        print("✓ No hay partidos nuevos para insertar.")

    # Recuperar catálogo completo (incluye los recién insertados)
    part_db = pd.read_sql(text("""
        SELECT partido_id,
               temporada,
               CAST(fecha AS date) AS fecha,
               equipo_local_id,
               equipo_visitante_id
        FROM dbo.Partidos
    """), conn)

# Emparejar para traer partido_id a df (alinear tipos antes del merge)
df_merge = df.copy()
df_merge["fecha"] = pd.to_datetime(df_merge["fecha"]).dt.date
part_db["fecha"] = pd.to_datetime(part_db["fecha"]).dt.date

key_cols = ["temporada", "fecha", "equipo_local_id", "equipo_visitante_id"]
df_merge = df_merge.merge(part_db, on=key_cols, how="left", validate="many_to_one")

if df_merge["partido_id"].isna().any():
    perdidos = df_merge[df_merge["partido_id"].isna()][key_cols].drop_duplicates()
    raise RuntimeError(f"No se pudo emparejar partido_id para algunas filas:\n{perdidos.head()}")

# 4) ESTADÍSTICAS 
def filas_stats(row):
    out = []
    # Local
    loc = {"partido_id": row["partido_id"], "equipo_id": row["equipo_local_id"], "es_local": 1}
    for xls_col, sql_col in STATS_LOCAL_XLS_TO_SQL.items():
        if xls_col in row.index:
            loc[sql_col] = pd.to_numeric(row[xls_col], errors="coerce")
    out.append(loc)
    # Visitante
    vis = {"partido_id": row["partido_id"], "equipo_id": row["equipo_visitante_id"], "es_local": 0}
    for xls_col, sql_col in STATS_VISITA_XLS_TO_SQL.items():
        if xls_col in row.index:
            vis[sql_col] = pd.to_numeric(row[xls_col], errors="coerce")
    out.append(vis)
    return out

stats_rows = []
for _, r in df_merge.iterrows():
    stats_rows.extend(filas_stats(r))

stats_df = pd.DataFrame(stats_rows)

# Casts suaves (manteniendo NULLs)
for c in ["tiros_totales","tiros_porteria","atajadas_portero","pases","saques_esquina","faltas","tarjetas_amarillas","tarjetas_rojas"]:
    if c in stats_df.columns:
        stats_df[c] = pd.to_numeric(stats_df[c], errors="coerce").astype("Int64")

# Evitar duplicados de estadística si ya existen combinaciones (partido_id, equipo_id)
with ENGINE.begin() as conn:
    existe_stats = pd.read_sql(text("SELECT partido_id, equipo_id FROM dbo.Estadisticas_partidos"), conn)

if not existe_stats.empty:
    stats_df = stats_df.merge(existe_stats.drop_duplicates(), on=["partido_id","equipo_id"], how="left", indicator=True)
    stats_df = stats_df[stats_df["_merge"] == "left_only"].drop(columns="_merge")

with ENGINE.begin() as conn:
    if not stats_df.empty:
        stats_df.to_sql("Estadisticas_partidos", conn, schema="dbo", if_exists="append", index=False)
        print(f"✓ Estadísticas insertadas: {len(stats_df)}")
    else:
        print("✓ No hay estadísticas nuevas para insertar.")

print("✅ Carga completada")
```
